````python
def evaluate_formula(formula):
    """
    Evaluates a nested formula string using predefined operations.
    Example: evaluate_formula("divide(subtract(206588, 181001), 181001)")
    """
    # Define the allowed operations
    def add(x, y): return x + y
    def subtract(x, y): return x - y
    def multiply(x, y): return x * y
    def divide(x, y): return x / y if y != 0 else float('inf')
    def exp(x, y): return x ** y
    def greater(x, y): return 1 if x > y else 0
    
    # Dictionary of available operations
    operations = {
        'add': add,
        'subtract': subtract,
        'multiply': multiply,
        'divide': divide,
        'exp': exp,
        'greater': greater
    }
    
    def parse_and_evaluate(s):
        """Recursively parses and evaluates the formula string."""
        # If it's a number, return it
        try:
            return float(s)
        except ValueError:
            pass
        
        # Find the first operation
        first_paren = s.find('(')
        if first_paren == -1:
            raise ValueError("Invalid formula format")
        
        op_name = s[:first_paren]
        if op_name not in operations:
            raise ValueError(f"Unknown operation: {op_name}")
        
        # Extract the arguments string
        args_str = s[first_paren + 1:-1]  # Remove outer parentheses
        
        # Parse arguments
        args = []
        paren_count = 0
        current_arg = ""
        
        for char in args_str:
            if char == '(' and paren_count >= 0:
                paren_count += 1
                current_arg += char
            elif char == ')' and paren_count > 0:
                paren_count -= 1
                current_arg += char
            elif char == ',' and paren_count == 0:
                args.append(current_arg.strip())
                current_arg = ""
            else:
                current_arg += char
        
        args.append(current_arg.strip())
        
        if len(args) != 2:
            raise ValueError(f"Operation {op_name} requires exactly 2 arguments")
        
        # Recursively evaluate arguments
        evaluated_args = [parse_and_evaluate(arg) for arg in args]
        
        # Apply the operation
        return operations[op_name](*evaluated_args)
    
    try:
        return parse_and_evaluate(formula)
    except Exception as e:
        raise ValueError(f"Error evaluating formula: {str(e)}")

# Test cases
test_cases = [
    "add(2, 3)",
    "multiply(subtract(10, 5), 3)",
    "divide(subtract(206588, 181001), 181001)",
    "exp(2, 3)",
    "greater(5, 3)"
]

for test in test_cases:
    try:
        result = evaluate_formula(test)
        print(f"{test} = {result}")
    except ValueError as e:
        print(f"Error in {test}: {str(e)}")
````
